## 游멆잺 Guardar y Restaurar la Configuraci칩n por Defecto en Matplotlib

Cuando trabajamos con estilos en Matplotlib usando `plt.style.use()`, es importante recordar que los cambios afectan **todos los gr치ficos generados a partir de ese momento**.  
Para evitar que un estilo modifique permanentemente la apariencia de nuestras visualizaciones, podemos seguir estos pasos:

1. **Guardar la configuraci칩n predeterminada de Matplotlib al inicio**.
2. **Aplicar el estilo deseado para un gr치fico espec칤fico**.
3. **Restaurar la configuraci칩n original despu칠s de usar el estilo**.

---

### 游댃 Ejemplo pr치ctico

```python
import matplotlib.pyplot as plt

# 1. Guardamos la configuraci칩n por defecto
IPython_default = plt.rcParams.copy()

# 2. Aplicamos un estilo predefinido
plt.style.use('fivethirtyeight')

# 3. Creamos el gr치fico con el estilo aplicado
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['A침o'], datos_col['Inmigrantes'])
ax.set_title('Inmigraci칩n de colombianos hacia Canad치\n1980 a 2013', fontsize=20, loc='left')
ax.set_ylabel('N칰mero de Inmigrantes', fontsize=14)
ax.set_xlabel('A침o', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()

# 4. Restauramos la configuraci칩n original
plt.rcParams.update(IPython_default);


### 游늷 Tip adicional

Puedes aplicar un estilo **temporalmente** a un bloque de c칩digo utilizando `with plt.style.context(...)`. Esto es 칰til para que el estilo solo afecte a los gr치ficos dentro de ese bloque:

```python
with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(datos_col['A침o'], datos_col['Inmigrantes'])
    ax.set_title('Inmigraci칩n de colombianos hacia Canad치\n1980 a 2013', fontsize=20, loc='left')
    ax.set_ylabel('N칰mero de Inmigrantes', fontsize=14)
    ax.set_xlabel('A침o', fontsize=14)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    plt.show()


### 九 Ventajas de este enfoque

- 游빞 **Aislamiento de estilos**: evita que el estilo se propague a todo el notebook o script.
- 游빍 **Experimentaci칩n controlada**: permite probar diferentes estilos sin alterar otros gr치ficos.
- 游 **Preparaci칩n profesional**: 칰til para crear gr치ficos con distintos formatos para informes, presentaciones o publicaciones.

````python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which='both', size=0)

  ax = sns.barplot(data=top_10, x='Total', y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canad치\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('N칰mero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  for i, j in enumerate(top_10['Total']):
    ax.text(j + 20, i, str(j), color='black', fontsize=9, ha='left',va='center')


  plt.show()
````

````python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])

  ax = sns.barplot(data=top_10, x='Total', y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canad치\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('N칰mero de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i, j in enumerate(top_10['Total']):
    ax.text(j + 20, i, str(j), color='black', fontsize=9, ha='left',va='center')


  plt.show()
````

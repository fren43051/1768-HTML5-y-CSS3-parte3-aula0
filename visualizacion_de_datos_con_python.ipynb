## 🛠️ Guardar y Restaurar la Configuración por Defecto en Matplotlib

Cuando trabajamos con estilos en Matplotlib usando `plt.style.use()`, es importante recordar que los cambios afectan **todos los gráficos generados a partir de ese momento**.  
Para evitar que un estilo modifique permanentemente la apariencia de nuestras visualizaciones, podemos seguir estos pasos:

1. **Guardar la configuración predeterminada de Matplotlib al inicio**.
2. **Aplicar el estilo deseado para un gráfico específico**.
3. **Restaurar la configuración original después de usar el estilo**.

---

### 🔄 Ejemplo práctico

```python
import matplotlib.pyplot as plt

# 1. Guardamos la configuración por defecto
IPython_default = plt.rcParams.copy()

# 2. Aplicamos un estilo predefinido
plt.style.use('fivethirtyeight')

# 3. Creamos el gráfico con el estilo aplicado
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.set_title('Inmigración de colombianos hacia Canadá\n1980 a 2013', fontsize=20, loc='left')
ax.set_ylabel('Número de Inmigrantes', fontsize=14)
ax.set_xlabel('Año', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()

# 4. Restauramos la configuración original
plt.rcParams.update(IPython_default);


### 📌 Tip adicional

Puedes aplicar un estilo **temporalmente** a un bloque de código utilizando `with plt.style.context(...)`. Esto es útil para que el estilo solo afecte a los gráficos dentro de ese bloque:

```python
with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
    ax.set_title('Inmigración de colombianos hacia Canadá\n1980 a 2013', fontsize=20, loc='left')
    ax.set_ylabel('Número de Inmigrantes', fontsize=14)
    ax.set_xlabel('Año', fontsize=14)
    ax.yaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_tick_params(labelsize=12)
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    plt.show()


### ✅ Ventajas de este enfoque

- 🧼 **Aislamiento de estilos**: evita que el estilo se propague a todo el notebook o script.
- 🧪 **Experimentación controlada**: permite probar diferentes estilos sin alterar otros gráficos.
- 🧾 **Preparación profesional**: útil para crear gráficos con distintos formatos para informes, presentaciones o publicaciones.

````python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10 = top_10.set_index('Pais')
top_10
````

````python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set(title='Top 10 paises con mayor migracion hacia Canadá\n1980 - 2013',
       xlabel='Número de inmigrantes',
       ylabel='Año')
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Top 10 paises con mayor migracion hacia Canadá\n1980 - 2013',
       loc='left', fontsize=16)
ax.set_xlabel('Número de Inmigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
````

````python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which='both', size=0)

  ax = sns.barplot(data=top_10, x='Total', y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canadá\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  for i, j in enumerate(top_10['Total']):
    ax.text(j + 20, i, str(j), color='black', fontsize=9, ha='left',va='center')


  plt.show()
````

````python
def generar_grafico(palette):
  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])

  ax = sns.barplot(data=top_10, x='Total', y= top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top 10 paises con mayor migracion hacia Canadá\n1980 - 2013',
        loc='left', fontsize=16)
  ax.set_xlabel('Número de Inmigrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i, j in enumerate(top_10['Total']):
    ax.text(j + 20, i, str(j), color='black', fontsize=9, ha='left',va='center')


  plt.show()
````

# 📊 Visualización interactiva con Plotly

Plotly es una biblioteca de Python para la creación de gráficos interactivos y altamente personalizables, ideal para visualización de datos moderna en la web o notebooks.

---

## 🚀 ¿Por qué usar Plotly?

- ✨ **Interactividad:** Zoom, desplazamiento, selección y tooltips integrados.
- 📱 **Responsivo:** Compatible con navegadores y dispositivos móviles.
- 🔌 **Integración:** Funciona con `pandas`, `NumPy`, `Dash`, y Jupyter.
- 🧩 **Soporte 3D y Mapas:** Incluye gráficos 3D, mapas geográficos, de calor, etc.

https://plotly.com/python/

---
